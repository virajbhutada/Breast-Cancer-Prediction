## ***3. Data Wrangling***

### Data Wrangling Code

```python
# Print all columns name
df.columns.tolist()
```

```python
# Rename some necessary columns name
df.rename(columns={'iyear':'year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'killeed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                       'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
# Updated dataset after rename some column name
df.columns
```

```python
# Fill missing values in the 'Wounded' column with zeros and convert the data type to integer
df['Wounded'] = df['Wounded'].fillna(0).astype(int)

# Fill missing values in the 'killeed' column with zeros and convert the data type to integer
df['killeed'] = df['killeed'].fillna(0).astype(int)

# Calculate the total casualties by adding the 'killeed' and 'Wounded' columns
df['casualities'] = df['killeed'] + df['Wounded']
```

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1 - Terrorist attacks in different years

```python
# Chart - 1 Visualization code for attacks in different years

# Create subplots with a specific size
plt.subplots(figsize=(13, 6))

# Create a count plot to visualize the number of terrorist activities each year
sns.countplot(x='year', data=df, palette='RdYlGn_r', edgecolor=sns.color_palette("YlOrBr", 5))

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Set the title of the graph
plt.title('Number of Terrorist Activities Each Year')

# Display the plot
plt.show()
```

##### What is/are the insight(s) found from the chart?

The chart reveals that the highest number of terrorist attacks occurred between 2012 and 2017. Among these years, 2014 witnessed the highest number of terrorist attacks.

#### Chart - 2 - Terrorist Activities by Region Over the Years

```python
# Chart - 2 Visualization code for attacks by region in different years
# Create a cross-tabulation of the number of terrorist attacks by year and region
cross_tab = pd.crosstab(df['year'], df['Region'])

# Calculate the sum of attacks in each region and sort the columns accordingly
cross_tab = cross_tab[cross_tab.sum().sort_values(ascending=False).index]

# Plot the cross-tabulation as an area plot
cross_tab.plot(kind='area', figsize=(12, 5))

# Set the title of the graph
plt.title('Terrorist Activities by Region Over the Years')

# Label the y-axis
plt.ylabel("Number of Attacks")

# Show the plot
plt.show()
```

##### What is/are the insight(s) found from the chart?

The graph reveals two significant findings:

- Middle East & North Africa consistently tops the list for terrorist activities, indicating a sustained high level of incidents.
- South Asia follows closely in terms of the number of incidents, signifying a noteworthy regional challenge.

#### Chart - 3 - Terrorist Incidents by City/State

```python
# Chart - 3 Visualization code for terrorist incidents by city/state
# Import WordCloud library
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Filter and get a Series of non-null state values from the DataFrame
cities = df['state'].dropna()

# Create a word cloud figure with specified dimensions and background color
plt.figure(figsize=(10, 10))
wordcloud = WordCloud(background_color='white', width=512, height=384).generate(' '.join(cities))

# Set the title of the graph
plt.title('Word Cloud of Terrorist Incidents by City/State')

# Turn off axis labels
plt.axis('off')

# Display the word cloud
plt.imshow(wordcloud)
plt.show()
```

##### What is/are the insight(s) found from the chart?

The word cloud highlights key regions in the context of terrorist incidents:

- **Northern Ireland:** A region with a historical association with terrorism.
- **Khyber Pakhtunkhwa and Tribal Areas:** Signifying vulnerability to terrorism in Pakistan.
- **Baghdad and Al Anbar:** Prominent locations in Iraq with a history of terrorist activities.

These regions have experienced a higher concentration of terrorism, warranting attention for security and stability efforts.

#### Chart - 4 - Top Countries Affected by Terrorist Attacks

```python
# Chart - 4 Visualization code for Top 20 Countries Affected by Terrorist Attacks
# Get the top 20 countries with the highest counts of terrorist attacks from the DataFrame
attack_country = df['Country'].value_counts()[:20]

# Create a subplots figure
plt.subplots(figsize=(10, 5))

# Create a bar plot to visualize the count of terrorist attacks by country
sns.barplot(x=attack_country.index, y=attack_country.values, palette="crest")

# Set the title of the graph
plt.title('Top Countries Affected by Terrorist Attacks')

# Set labels
plt.xlabel('Countries')
plt.ylabel('Count of Attacks')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the bar plot
plt.show()
```

##### What is/are the insight(s) found from the chart?

The analysis of the top countries affected by terrorist attacks reveals the following:

- Iraq is the most severely impacted country, indicating ongoing security challenges.

- Pakistan ranks second, reflecting a significant presence of terrorism-related incidents due to its geopolitical context.

- Afghanistan follows as the third-most affected country, highlighting its vulnerability to terrorism.

- India is also among the top countries affected, emphasizing the need for addressing security concerns.

#### Chart - 5 - Top terrorist groups involved in the most attacks

```python
# Chart - 5 Visualization code for top terrorist groups involved in the most attacks
# Get the top 10 terrorist groups involved in the most attacks from the 'Group' column
attack_groups = df['Group'].value_counts()[1:10]

# Create a subplots figure with a specific size for the bar plot
plt.subplots(figsize=(10, 5))

# Create a bar plot to visualize the count of terrorist attacks by group
sns.barplot(x=attack_groups.values, y=attack_groups.index, palette="crest")

# Set the title of the graph.
plt.title('Terrorist Groups Involved Most in Terrorism')

# Set labels
plt.xlabel('Count')
plt.ylabel('Groups')

# Display the bar plot
plt.show()

# Print the 'attack_groups' data for further analysis
print("Top Terrorist Groups Involved in Terrorism (Count of Attacks):")
print(attack_groups)
```

#### Chart - 6 - Number of People Killed by Each Terrorist Group

```python
# Chart - 6 Visualization code for number of people killed by each terrorist group
# Calculate the total number of people killed by each terrorist group
group_killed = df[['Group', 'killeed']].groupby(['Group'], axis=0).sum().sort_values('killeed', ascending=False).head(10)

# Create a subplots figure with a specific size for the bar plot
plt.subplots(figsize=(10, 5))

# Create a bar plot using Seaborn to visualize the number of people killed by each terrorist group
sns.barplot(x=group_killed.killeed.values, y=group_killed.index, palette="crest")

# Set the title of the graph
plt.title('Number of People Killed by Each Terrorist Group')

# Set labels
plt.xlabel('Number of People Killed')
plt.ylabel('Terrorist Groups')

# Display the bar plot
plt.show()

# Print the 'group_killed' data for further analysis
print("Number of People Killed by Each Terrorist Group:")
print(group_killed)
```

##### What is/are the insight(s) found from the chart?

In the analysis of the number of people killed by terrorist groups, several key findings emerge:

- **Unknown:** Surprisingly, the category labeled as "Unknown" stands out as the top contributor to casualties, with a significant count of 112,367 individuals killed. This highlights the challenges in identifying responsible groups in many terrorism-related incidents.

- **Islamic State of Iraq and the Levant (ISIL):** ISIL follows closely, with 38,923 individuals killed, underlining their substantial role in inflicting harm.

#### Chart - 7 - Number of People Died in Each Year

```python
# Chart - 7 Visualization code for number of people died in each year
# Calculate the total number of people died in each year
died_people = df[['year', 'killeed']].groupby(['year']).sum()

# Create a subplots figure with a specific size for the bar plot
plt.subplots(figsize=(13, 6))

# Create a bar plot using Seaborn to visualize the number of people who died in each year
sns.barplot(x=died_people.index, y=died_people.killeed.values, palette="rocket_r")

# Set the title of the graph
plt.title("Number of People Died in Each Year", fontsize=15)

# Set labels
plt.ylabel("Number of People Died")
plt.xlabel('Year')

# Rotate x-axis labels for better readability.
plt.xticks(rotation=45)

# Display the bar plot.
plt.show()

# Print the 'died_people' data for further analysis.
print("Top 15 Years with the Highest Number of People Died:")
print(died_people.nlargest(15, 'killeed'))
```

##### What is/are the insight(s) found from the chart?

In summary, the analysis of the number of people who died in different years reveals several notable findings:

- The year 2014 witnessed the highest number of casualties, with 44,490 individuals losing their lives in terrorist incidents. This signifies a particularly deadly year in terms of terrorist violence.

- The year 2016 saw a considerable number of casualties, with 34,871 individuals losing their lives. Although the count decreased compared to the previous years, it remained a challenging year in terms of security.

- Following the peak years, 2017, 2013, and 2012 were also marked by significant casualties, indicating that the threat of terrorism remained a pressing global concern during these years.

#### Chart - 10 - Number of People Killed in Each Attack Type

```python
# Chart - 10 Visualization code for number of people killed in each attack type
# Calculate the total number of people killed in each attack type
attack_killed = df[['AttackType', 'killeed']].groupby(["AttackType"], axis=0).sum()
attack_killed = attack_killed.sort_values(by='killeed', ascending=False)  # Sort the data in descending order

# Create a subplots figure with a specific size for the bar plot
plt.subplots(figsize=(10, 5))

# Create a bar plot to visualize the number of people killed in each attack type
sns.barplot(x=attack_killed.killeed.values, y=attack_killed.index, palette="cubehelix")

# Set the title of the graph
plt.title('Number of People Killed in Each Attack Type')

# Set labels
plt.xlabel('Number of People Killed')
plt.ylabel('Type of Attacks')

# Display the bar plot
plt.show()

# Print the 'attack_killed' data for further analysis
print("Number of People Killed in Each Attack Type:")
print(attack_killed)
```

##### What is/are the insight(s) found from the chart?

The analysis of the number of people killed in various attack types reveals several significant findings:

- Armed assaults top the list with the highest number of casualties, totaling 160,297. This indicates the severe impact of direct armed confrontations in causing loss of life.

- Bombing and explosion attacks closely follow, with 157,321 casualties. These incidents, often involving explosive devices, result in a significant number of deaths.

- The "Unknown" category comes in the third position, suggesting challenges in identifying the exact nature of certain attacks that lead to casualties.

- Assassinations are another notable category, contributing to a substantial number of casualties.

- Incidents involving hostage taking and kidnapping also result in significant loss of life.

These insights underscore the different patterns of attack types and their associated human costs.

#### Chart - 11 - Number of People Wounded in Each Attack Type

```python
# Chart - 11 Visualization code for total number of people wounded in each attack type
# Calculate the total number of people wounded in each attack type
attack_wounded = df[['AttackType', 'Wounded']].groupby(["AttackType"], axis=0).sum()
attack_wounded = attack_wounded.sort_values(by='Wounded', ascending=False)  # Sort the data in descending order

# Create a subplots figure with a specific size for the bar plot
plt.subplots(figsize=(10, 5))

# Create a bar plot to visualize the number of people wounded in each attack type
sns.barplot(x=attack_wounded.Wounded.values, y=attack_wounded.index, palette="viridis")

# Set the title of the graph
plt.title('Number of People Wounded in Each Attack Type')

# Set labels
plt.xlabel('Number of People Wounded')
plt.ylabel('Type of Attacks')

# Display the bar plot
plt.show()

# Print the 'attack_wounded' data for further analysis
print("Number of People Wounded in Each Attack Type:")
print(attack_wounded)
```

##### What is/are the insight(s) found from the chart?

The analysis of the number of people wounded in various attack types reveals the following key findings:

- Bombing and explosion attacks are the leading cause of injuries, resulting in a significant count of 372,686 wounded individuals. These incidents often lead to a large number of casualties and injuries.

- Armed assaults come next with 77,365 wounded individuals, indicating that direct confrontations and the use of firearms also result in a considerable number of injuries.

These insights emphasize the severe physical and emotional toll that bombing/explosion and armed assault attacks can have. Understanding the patterns of injuries in various attack types is essential for developing strategies to prevent and mitigate the impact of terrorism.
